```python
import json

# JSON 파일 경로
with open("all_pattern_predicates.json", "r", encoding="utf-8") as file:
    pattern_predicates = json.load(file)

# 법률 매핑
law_mapping = {
    "comparison_prevention": "consumer_rights_law",
    "confirmshaming": "consumer_protection_law",
    "disguised_ads": "advertising_transparency_law",
    "fake_scarcity": "fair_trade_law",
    "fake_social_proof": "consumer_protection_law",
    "fake_urgency": "consumer_protection_law",
    "forced_action": "informed_consent_law",
    "hard_to_cancel": "cancellation_rights_law",
    "hidden_costs": "price_transparency_law",
    "hidden_subscription": "subscription_disclosure_law",
    "nagging": "user_interface_ethics_law",
    "obstruction": "accessibility_law",
    "preselection": "default_choice_fairness_law",
    "sneaking": "unfair_commercial_practices_law",
    "trick_wording": "clear_language_law",
    "visual_interference": "visual_design_ethics_law"
}

case_prefix = "case"

predicate_case_facts = []
predicate_type_facts = []
type_law_facts = []

# 데이터 처리
for i, (pattern, predicates) in enumerate(pattern_predicates.items(), start=1):
    case_id = f"{case_prefix}{str(i).zfill(3)}"
    law = law_mapping.get(pattern, "general_consumer_law")

    for predicate_call in predicates:
        predicate_name = predicate_call.split("(")[0]
        predicate_case_facts.append(f"predicate_case({predicate_name}, {case_id}).")
        predicate_type_facts.append(f"predicate_type({predicate_name}, {pattern}).")

    type_law_facts.append(f"type_law({pattern}, {law}).")

# 파일로 저장
with open("predicate_case.pl", "w", encoding="utf-8") as f:
    f.write("% predicate ↔ case\n")
    f.write("\n".join(predicate_case_facts))

with open("predicate_type.pl", "w", encoding="utf-8") as f:
    f.write("% predicate ↔ pattern\n")
    f.write("\n".join(predicate_type_facts))

with open("type_law.pl", "w", encoding="utf-8") as f:
    f.write("% pattern ↔ law\n")
    f.write("\n".join(type_law_facts))

print("✅ 모든 .pl 파일이 성공적으로 저장되었습니다.")
```
