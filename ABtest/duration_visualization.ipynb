```python
import pandas as pd
```

```python
# 시나리오 1 prev test -> 낯설음 / 보통 / 능숙함 그룹 나눔

df_s1 = pd.read_csv("../ABtest/editdata/scenario1_prev_test.csv")
df_s1
```

```python
# 분류 함수 정의
def classify_experience(value):
    if "1. 너무 낯설어서" in value or "2. 속도는 느리지만" in value:
        return "낯설음"
    elif "4. 낯설진 않지만" in value:
        return "보통"
    elif "5. 능숙하게" in value:
        return "능숙함"
    else:
        return "기타"

# 사용능숙도 컬럼 추가 
df_s1["사용능숙도"] = df_s1["키오스크를 어떻게 사용하고 계신가요?"].apply(classify_experience)
df_s1
# 결과 확인
df_s1.to_csv("../ABtest/editdata/scenario1_prev_test.csv",index=False)
```

```python
df_s1_clicklogs = pd.read_csv('../ABtest/editdata/scenario1_clicklogs.csv')

df_s1_clicklogs
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 자동 로드 (Windows 기준: Malgun Gothic)
font_path = fm.findSystemFonts(fontpaths=None, fontext='ttf')
malgun_font = [f for f in font_path if 'Malgun' in f]
if malgun_font:
    plt.rcParams['font.family'] = fm.FontProperties(fname=malgun_font[0]).get_name()
plt.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
df_s1 = pd.read_csv("../ABtest/editdata/scenario1_prev_test.csv")
df_s1_clicklogs = pd.read_csv("../ABtest/editdata/scenario1_clicklogs.csv")

# 사용능숙도 매핑 (성함 기준)
# 중복 제거 (사람당 하나의 사용능숙도만 있도록)
user_level = df_s1[["성함을 작성해주세요.", "사용능숙도"]].drop_duplicates()
user_level.columns = ["username", "사용능숙도"]

# 클릭로그와 매핑
df_merged = pd.merge(df_s1_clicklogs, user_level, on="username", how="left")

# total_duration_sec는 사람마다 고정이므로 한 사람당 1개만 남김
df_total = df_merged[["username", "total_duration_sec", "사용능숙도"]].drop_duplicates()

# 시각화
plt.figure(figsize=(10, 6))
for level in df_total["사용능숙도"].unique():
    sub = df_total[df_total["사용능숙도"] == level]
    plt.barh(sub["username"], sub["total_duration_sec"], label=level)

plt.xlabel("총 소요 시간 (초)")
plt.title("사용능숙도 그룹별 total_duration_sec")
plt.legend(title="사용능숙도")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 한글 폰트 설정 (Mac용)
plt.rcParams['font.family'] = 'Apple SD Gothic Neo'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
df_s1 = pd.read_csv("../ABtest/editdata/scenario1_prev_test.csv")
df_s1_clicklogs = pd.read_csv("../ABtest/editdata/scenario1_clicklogs.csv")

# 사용능숙도 매핑
user_level = df_s1[["성함을 작성해주세요.", "사용능숙도"]].drop_duplicates()
user_level.columns = ["username", "사용능숙도"]

# 클릭로그와 매핑
df_merged = pd.merge(df_s1_clicklogs, user_level, on="username", how="left")

# 총 소요 시간 (중복 제거)
df_total = df_merged[["username", "total_duration_sec", "사용능숙도"]].drop_duplicates()

# 평균값 계산
grouped = df_total.groupby("사용능숙도")["total_duration_sec"].mean().reindex(["능숙함", "보통", "낯설음"])

# 시각화
plt.figure(figsize=(8, 5))
grouped.plot(kind='bar', color=['green', 'orange', 'red'])

plt.title("사용능숙도 그룹별 평균 소요 시간")
plt.ylabel("평균 소요 시간 (초)")
plt.xlabel("사용능숙도")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
